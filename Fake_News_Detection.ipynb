## Defining a function called 'wordpot' that will preprocess all the texts in the dataframe before the detection can be done that the News is Fake or True.

```python
import re #Regular expression library.
import string

def wordpot(text):
  text = text.lower()
  text = re.sub('\[.*?\]', '', text)
  text = re.sub('\\W', " ", text)
  text = re.sub('https?://\S+|www\.\S+', '', text)
  text = re.sub('<.*?>+', '', text)
  text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
  text = re.sub('\n', ' ', text)
  text = re.sub('\w*\d\w*', '', text)
  return text
```

```python
df_merge["text"] = df_merge["text"].apply(wordpot)
```

```python
df_merge.isnull().sum()
```

## Now we will define the dependent and independent variable.

```python
X = df_merge["text"]
Y = df_merge["class"]
```

## Splitting the training and testing data. For this Fake news detection project, I will keep keep the test data as 25%.


```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.25)
```

Text to Vector conversion

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorization = TfidfVectorizer()
xv_train = vectorization.fit_transform(x_train)
xv_test = vectorization.transform(x_test)
```

Now, I will Apply Logistic Regression for the detection of Fake News.

```python
#Logistic Rgression
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_curve, auc

#Train the model
model = LogisticRegression()
model.fit(xv_train, y_train)

#Now I will Evaluate the Model
y_pred = model.predict(xv_test)

accuracy_score = accuracy_score(y_test, y_pred)
print("The accuracy score for the Logistic Regression model is:", accuracy_score)
```

```python
#Confusion Matrix for the Logistic Regression model.
print("Confusion Matrix:\n\n", confusion_matrix(y_test, y_pred))
```

```python
#The classification report for the LR model
print("Classification Report:\n\n", classification_report(y_test, y_pred))
```

## Decision Tree Classification
